```python
import numpy as np
import scipy
import sklearn
import matplotlib.pyplot as plt
```

```python
def alpha_bar(x):
    return 1/8(x + 1)**2

def eigen_generator(n):
    # n >= 1
    return 1 / ((n-0.5)**2 * np.pi)

def eigen_func(x, eigen_value):
    # n >= 1
    return np.sqrt(2)*np.sin(x / np.sqrt(eigen_value))

eigen_func = np.vectorize(eigen_func)
```

```python
first_3_eigens  = eigen_generator(np.arange(1,4))

x_domain = np.linspace(-1,1,201)
y_values = np.zeros((201, 3))

for i in range(3):
    y_values[:,i] = eigen_func(x_domain, first_3_eigens[i])
    
```

```python
fig, ax = plt.subplots(ncols=1, figsize=(8, 4))

for i in range(3):
    plt.plot(x_domain, y_values[:,i], label=f'Eigenfunction {i+1}')
plt.legend()
plt.title(r'$c(x,y)= {\mathrm{min}}(x,y)$');
```
